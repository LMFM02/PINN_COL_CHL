```python
import numpy as np
import matplotlib.pyplot as plt
```

# **Compuerta CNOT**

```python
#Definimos la escala de tiempo
times=np.linspace(0.0,5.0,500)
```

## **Acople $\sigma_x\otimes \sigma_x$**

```python
#Extraemos el archivo binario para este acople
sweep_xx=np.load("../Results/CNOT_lr1e-04-1e-03_neurons100-500_Hxx.npy", allow_pickle=True)
```

### **Mejor escenario (máxima fidelidad promedio)**

```python
#Extraemos todas las fidelidades promedio
fidelities_xx = np.array([d["A. fidelity"] for d in sweep_xx])

#Extraemos el indice que tiene la mayor fidelidad promedio 
id_xx = np.argmax(fidelities_xx)

#Mejor fidelidad promedio
best_fidelity_xx = fidelities_xx[id_xx]
print("Mejor fidelidad promedio: ", best_fidelity_xx)
print("Indice:", id_xx)
```

```python
#Valor de learning rate y neuronas, que generan el mejor entrenamiento
lr_xx=sweep_xx[id_xx]["learning rate"]
n_xx=sweep_xx[id_xx]["neurons"]

print("learning rate: ", lr_xx)
print("neurons: ", n_xx)
```

```python
#########################################################################
#Graficamos la perdida para el entrenamiento con mejor fidelidad promedio
#########################################################################

fig_loss = plt.figure(figsize=(600/80, 300/80))  # Dividido por 80 para convertir píxeles a pulgadas (80 píxeles por pulgada es común)
ax_loss = fig_loss.add_subplot(111)
plt.yscale("log")
#ax_loss.set_title('Términos de Pérdida', fontsize=16)
ax_loss.set_xlabel('Epochs (log scale)')
ax_loss.set_ylabel('Loss (log scale)')
ax_loss.loglog(sweep_xx[id_xx]["Loss"], 'b-', label=f"Total Loss Function | Min Loss: {sweep_xx[id_xx]["Loss"][-1]:.2e}")
ax_loss.loglog(sweep_xx[id_xx]["L_model"], 'r', label=f"Loss Model          | Min Loss: {sweep_xx[id_xx]["L_model"][-1]:.2e}")
ax_loss.loglog(sweep_xx[id_xx]["L_control"], 'lime', label=f"Loss Control        | Min Loss: {sweep_xx[id_xx]["L_control"][-1]:.2e}")
ax_loss.loglog(sweep_xx[id_xx]["L_const"], 'cyan', label=f"Loss Const          | Min Loss: {sweep_xx[id_xx]["L_const"][-1]:.2e}")
ax_loss.loglog(sweep_xx[id_xx]["L_reg"], 'purple', label=f"Loss Reg            | Min Loss: {sweep_xx[id_xx]["L_reg"][-1]:.2e}")
ax_loss.grid(True)
ax_loss.legend(prop={'family': 'monospace', 'size': 9})
```

```python
#########################################################################
#Graficamos los controles del mejor entrenamiento
#########################################################################

fig, ax =plt.subplots(3,2, figsize=(8,5))

ax[0,0].plot(times, sweep_xx[id_xx]["ex1"], 'b', linestyle='solid', linewidth=2, label=r'$u_{x}^{(1)}(t)$')
ax[0,0].legend(fontsize=12)
ax[0,0].grid()

ax[0,1].plot(times, sweep_xx[id_xx]["ex2"], 'k', linestyle='solid', linewidth=2, label=r'$u_{x}^{(2)}(t)$')
ax[0,1].legend(fontsize=12)
ax[0,1].grid()

ax[1,0].plot(times, sweep_xx[id_xx]["ey1"], 'r', linestyle='solid', linewidth=2, label=r'$u_{y}^{(1)}(t)$')
ax[1,0].legend(fontsize=12)
ax[1,0].grid()

ax[1,1].plot(times, sweep_xx[id_xx]["ey2"], 'm', linestyle='solid', linewidth=2, label=r'$u_{y}^{(2)}(t)$')
ax[1,1].legend(fontsize=12)
ax[1,1].grid()

ax[2,0].plot(times, sweep_xx[id_xx]["ez1"], 'g', linestyle='solid', linewidth=2, label=r'$u_{z}^{(1)}(t)$')
ax[2,0].legend(fontsize=12)
ax[2,0].set_xlabel('Time [a.u]', fontsize=12)
ax[2,0].grid()

ax[2,1].plot(times, sweep_xx[id_xx]["ez2"], 'c', linestyle='solid', linewidth=2, label=r'$u_{z}^{(2)}(t)$')
ax[2,1].set_xlabel('Time [a.u]', fontsize=12)
ax[2,1].legend(fontsize=12)
ax[2,1].grid()

fig.supylabel('Control functions', fontsize=14, x=0.04)
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

```python
#Extraemos todos los valores de learning rate y neuronas del barrido
learning_rate_xx= [d["learning rate"] for d in sweep_xx]   
neuronas_xx= [d["neurons"] for d in sweep_xx]
```

```python
##################################################################################
#Graficamos una superficie tipo 3D de learning rate, neuronas y fidelidad promedio
##################################################################################

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(16, 8))

# Graficar puntos 3D
sc = ax.scatter(learning_rate_xx, neuronas_xx, fidelities_xx, 
                c=fidelities_xx, cmap='winter', s=100)

# Barra de color
fig.colorbar(sc, ax=ax, shrink=0.6, label='Average fidelity')

# Etiquetas
ax.set_ylabel('Neurons')
ax.set_xlabel('Learning rate')

plt.show()
```
